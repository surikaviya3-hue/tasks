## 3) standandize payment_mode and calculate the percentage share of each mode per city

```python
df['payment_mode']=df['payment_mode'].str.strip().str.lower().str.title()
df['city']=df['city'].str.strip().str.lower().str.title()
```

```python
df['payment_mode']
```

```python
df['city']
```

```python
df['payment_mode'].value_counts()
```

```python
df['city'].value_counts()
```

```python
share=(df.groupby(['payment_mode','city']).size())
share
```

```python
percentage=share/share.groupby(level=0).sum() * 100
percentage
```

```python
percentage_df = pd.DataFrame(percentage).reset_index()
percentage_df
```
