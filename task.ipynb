## 4)clean user_id by replacing empty strings with  nan and converting values to uppercase

```python
import numpy as np
df["user_id"]=df["user_id"].replace("",np.nan).str.upper()
df["user_id"]
```

## 5)convert age to numeric and replace invalid values(negative or >100) with NaN.

```python
df["age"]=pd.to_numeric(df["age"],errors="coerce")
```

```python
df.loc[(df["age"]<0) | (df["age"]>100),"age"]=np.nan
```

```python
df["age"]
```

## 6)clean salary by removing currency symbols and converting it to numeric.

```python
df
```

```python
df.dtypes
```

```python
df["salary"]=df["salary"].astype(str).str.replace("₹","",regex=False).replace("N/A",np.nan)
df["salary"]=pd.to_numeric(df["salary"],errors="coerce")
```

```python
df["salary"]
```

 ## 7)convert join_date into datetime format handling multiple date formats.

```python
df["join_date"]=pd.to_datetime(df["join_date"],format="%m-%d-%y",errors="coerce",dayfirst=True)
```

## 8)extract joining year from the cleaned join date.

```python
df["joining_year"]=df["join_date"].dt.year
```

## 11)convert performance score into numeric by mapping text values to numbers.

```python
df["performance_score"].unique()
```

## 12)clean experience_years by converting text values to numeric and removing negatives.

```python
df["experience_years"].unique()
```

```python
m={'five':5}
df["experience_years"]=df["experience_years"].replace(m)
```

```python
df["experience_years"]=pd.to_numeric(df["experience_years"],errors='coerce')
```

## 13)convert is_active values(yes/no/True/False) into boolean format.

```python
df["is_active"]
```

```python
df["is_active"].unique
```
