## 5. Clean salary and detect outliers using the IQR method. 

```python
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = df[
    (df['salary'] < lower_bound) |
    (df['salary'] > upper_bound)
]

outliers
```

## 15. Build a validation rule to flag rows violating at least two business constraints

```python
c1 = df['age'].isna() | (df['age'] < 18) | (df['age'] > 70)
c2 = df['salary'].isna() | (df['salary'] <= 0)
c3 = df['performance_rating'].isna()
c4 = (df['joining_date'] > df['last_promotion_date'])
c5 = (df['experience_years'] > (df['age'] - 18))
```

```python
df['violation_count'] = c1.astype(int) + c2.astype(int) + c3.astype(int) + c4.fillna(False).astype(int) + c5.fillna(False).astype(int)
flagged_rows = df[df['violation_count'] >= 2]
flagged_rows
```
